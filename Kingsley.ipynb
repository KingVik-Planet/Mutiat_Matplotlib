```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["Built Area", "Water", "Bare Soil", "Vegetation", "Hill"])
y = np.array([200488, 197780, 53938,92802,493048])

plt.bar(x,y, color = "cyan", width = 0.3, align = "edge")
plt.title("2023 Land Cover CLasses")
plt.xlabel("CLasses")
plt.ylabel("Area (Ha)")
plt.show()
```
